```python
import pandas as pd
import kagglehub

# Download the dataset locally using Kaggle Hub
dataset_path = kagglehub.dataset_download("sazidthe1/sustainable-development-report")

# Now read the specific CSV file using pandas
file_path = f"{dataset_path}/sustainable_development_report_2023.csv"
df = pd.read_csv(file_path)

# 1. Get general information about the dataset
print("=== Data Info ===")
print(df.info())

# 2. Get summary statistics for numerical columns
print("\n=== Data Description ===")
print(df.describe())

# 3. Check for missing values in each column
print("\n=== Missing Values ===")
print(df.isnull().sum())
```
